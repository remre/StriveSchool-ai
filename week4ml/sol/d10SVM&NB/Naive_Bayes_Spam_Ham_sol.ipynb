In this project you need to classify the emails as either 'Spam' (1) or 'Not Spam' (0).

You can notice that the emails have not the form of a simple text: they have as columns all the possible words you can find in the corpus of emails and as value of the column there is how many times that word occurs in the email.

When a spam classifier looks at an email, it searches for potential words that it has seen in the previous spam emails. If it finds a majority of those words, then it labels it as 'Spam'. 

Example 1 : let's take a word 'Greetings'. Say, it is present in both 'Spam' and 'Not Spam' emails.

Example 2 : Let's consider a word 'lottery'. Say, it is present in only 'Spam' emails.

Example 3 : Let's consider a word 'cheap'. Say, it is present only in 'Spam' emails.

If now we get a test email, and it contains all the three words metioned above, there's high probability that it is a 'Spam' mail.

The most effective algorithm for text-classification problems is the Naive-Bayes algorithm, that works on the classic Bayes' theorem. This theorem works on every individual word in the test data to make predictions(the conditional probability with higher probability is the predicted result).

From Wikipedia:

Let's suppose the suspected message contains the word "replica". Most people who are used to receiving e-mail know that this message is likely to be spam, more precisely a proposal to sell counterfeit copies of well-known brands of watches. The spam detection software, however, does not "know" such facts; all it can do is compute probabilities.
The formula used by the software to determine that, is derived from Bayes' theorem:

$$Pr(S|W) = \frac{Pr(W|S)\dot Pr(S)}{Pr(W|S)\dot Pr(S)+Pr(W|H)\dot Pr(H)}$$

where:

- Pr(S|W) is the probability that a message is a spam, knowing that the word "replica" is in it;

- Pr(S) is the overall probability that any given message is spam;

- Pr(W|S) is the probability that the word "replica" appears in spam messages;
- Pr(H) is the overall probability that any given message is not spam (is "ham");

- Pr(W|H) is the probability that the word "replica" appears in ham messages.

Read more on: https://en.wikipedia.org/wiki/Naive_Bayes_classifier

```python
#gausssian look solve it better
gnb = GaussianNB()
clf = MultinomialNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
y_predc = clf.fit(X_train, y_train).predict(X_test)

print("Number of mislabeled points out of a total %d points : %d"
      % (X_test.shape[0], (y_test != y_predc).sum()))
```
