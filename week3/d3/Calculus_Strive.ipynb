##Exponents



The exponent of a number says how many times to use the number in a multiplication.

*Example*:
if we need to calculate a^3 we need to multiply a with 3 times 
a^3 = a x a x a 


!!During the coding be careful with using **paranthesis** and consider **Order of operations**  always¡¡

!!!!!WATCH OUT!!!!

(a^2)^3 != (a^2)x(a^3)

```python
""" How to execute exponents in python"""
#The function will take the nth power of x 
def expo(x,n):
  x = (x**n)  #use "**" operator when then specify the power of exponent!! 
  return x
expo(2,3)
```

```python
#or you can use "pow()" function to execute exponents, see how it is done

pow(4,3)      # 4 is stands for base value and 3 is stands for power of exponent 
```

##Derivatives




The whole idea of differentiation is so central to calculus that
there is a special notation for it.
First, the Greek letter δ, i.e. ‘delta’, denotes not a number but
the phrase ‘increase in . . .’. So, for example, if x were to increase
from 1 to 1.01, then δx would be 0.01.
In this way, then, δx and δy denote the small increases in
x and y that occur as we move along some curve from the
point P to a nearby point Q.


image.png

Now, as we have seen, the whole process of differentiation
involves finding the limit of δy/δx as δx → 0, i.e. as Q moves closer
and closer to P.
And we now denote this limit by the special symbol dy/dx.

image.png

*Example:*

Derivative of f(x) = x**2


image.png

Now offer a second example, if only to show the dy/dx
notation in action.
Suppose, then, that y = 1/x.
Notably, as x increases, y decreases in this case,  so we might reasonably expect a negative slope,
and therefore a negative value of dy/dx.

###How to make it in Python

```python
diff(cos(x),x) #derivative of cos(x) with respect to x
```

```python
diff(y,x,2)
diff(y,x,x)# take the derivative of "y" with respect to "x" power of 2 this is how to do it in 2 ways
```
