###Polynomial Derivatives

```python
from sympy import *
x = Symbol("x")
y = Symbol("y")
diff(x**2+y**2,x)
```

```python
diff(x**2+y**2,y)
```

```python
diff(x**2*y**2,y)
```

```python
diff(x**2/y**2,x)
```

```python
z = (x**2+2*x)/(3*x+y**3)
diff(z,y)
```

###Trigonometry

```python
diff(cos(x*y),y)
```

```python
diff(sin(x*y),y)
```

```python
diff(cos(x**2+4*y),y)
```

```python
diff(sin(sqrt(x/y)),x)
```

```python
diff(sin((x**1/2)/(y**1/2)),x)
```
