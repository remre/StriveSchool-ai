```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

data = pd.read_csv("company_sales_data.csv")
plt.xlim(1,12)
plt.plot(data['month_number'],data['total_profit'])
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Company Profit Per Month')
```

```python
plt.figure(figsize=[10,8])
plt.xlim(0,12.5)
plt.plot(data['month_number'],data['total_profit'],color='r', marker='o',linestyle='--',linewidth=3,markersize=6, markerfacecolor='black')
plt.xlabel('Month Number')
plt.ylabel('Profit')
plt.title('Company Sales Data Of Last Year')
```

```python
data.head()
```

```python
fig, ax = plt.subplots(figsize=(10,8))

products = [['facecream','b','facecream Sales Data'],
    ['facewash','c','facewash Sales Data'],
    ['toothpaste','g','toothpaste Sales Data'],
    ['bathingsoap','r','bathingsoap Sales Data'],
    ['shampoo','m','shampoo Sales Data'],
    ['moisturizer','k','moisturizer Sales Data']]

for product in products:
    plt.plot(data['month_number'],data[product[0]],color=product[1],label=product[2],marker='o',linewidth=3)
    
plt.xlabel('month numbers')
plt.ylabel('sales units in numbers')
plt.title('Sales Data')
plt.legend()
```

```python
fig = plt.subplots(figsize=(10,8))
plt.bar(data['month_number']-0.2,data['facecream'],align='center',width=0.4)
plt.bar(data['month_number']+0.2,data['facewash'],align='center',width=0.4)



plt.grid(linestyle='--',linewidth=3)
plt.xlabel('Month Number', fontsize=20)
plt.ylabel('Number of units sold', fontsize=20)
plt.title('Facewash and Facecream Sales Data',fontsize=30)

colors = {'Facecream':'blue', 'Facewash':'orange'}  
labels = list(colors.keys())

handles = [plt.Rectangle((1,1),0,0, color=colors[label]) for label in labels]
plt.legend(handles, labels)
```
